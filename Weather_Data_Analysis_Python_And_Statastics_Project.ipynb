```python
def fill_null(x):
    for i in x.columns:
        if x[i].dtypes=="object":
            x[i]=x[i].fillna(x[i].mode()[0])
        elif x[i].dtypes != "object":
            x[i]=x[i].fillna(x[i].mean())
    
```
